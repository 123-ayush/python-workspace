```python
# ---------------------------------Constructor-------------------------------------------

# __init__ method

class car:
    def __init__(self, name):
        self.name = name
        print("constructor called...")

    def get_name(self):
        return self.name

car1 = car("bmw")
car2 = car("audi")

print(car1.get_name())
print(car2.get_name())


# default constructor : constructor having only self parameter.
# parameterized constructor : constructor having self along with other parameters.

```

```python
# -----------------------------------------Attributes------------------------------------------

# class attributes : attri. which are common for all objects

# instance attributes: attri. which are unique for each object

class bike:
    model = "two-wheeler" # class attribute

    def __init__(self, name, engine_cc):
        self.name = name #instance attr.
        self.engine_cc - engine_cc # instance attr.
```

```python
# --------------------------------Methods-------------------------------------------------------

# 1. Instance methods -> must contain self as parameter
#                     -> can access both class and instance attri.

# 2. Class methods -> must contain cls as parameter
#                  -> can access only class parameters
#                  -> also has decorator '@classmethod' which is used just before method to make that method class method

# 3. Static methods -> no compulsory parameter
#                   -> neither access class nor instance attrib.
#                  -> also has decorator '@staticmethod' which is used just before method to make that method static method
```

```python
class product:
    prod_Count = 0

    def __init__(self, name, price):
        self.name = name
        self.price = price
        product.prod_Count += 1

    @staticmethod
    def discounted_price(price, discount):
        final_price = price - (price * discount/100)
        return final_price
    
    @classmethod
    def get_count(cls):
        print(cls.prod_Count)


prod1 = product("soap", 100)
prod1 = product("mobile", 10000)
prod1 = product("laptop", 100000)
prod1 = product("computer", 1000000)


print(prod1.discounted_price(prod1.price, 10))
product.get_count()
```
