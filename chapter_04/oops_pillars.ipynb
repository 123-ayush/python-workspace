```python
# -----------------------------------------Encapsulation-----------------------------------------

# -> binding data and methods together in a class/single unit is called encapsulation

# --------------------------------------data hiding---------------------------------------------

# 1. public attr. -> accessible everywhere in code
# 2. protected attr. -> to make any attr protected we need to put underscore before identifier name (inside class + subclass)
# 3. private attr. -> to make any attr protected we need to put double underscore before identifier name (inside class only)
# to access private attr. we need to use getter and setters
# to access private attr outside use this: 'obj._className__attrName'
```

```python
# ----------------------------------------Inheritance--------------------------------------------

# reusing methods/attr. of parent class in derived class

# Types:

    # 1. Single level
    # 2. Multilevel
    # 3. Multiple (class1, class2, ..........)
    # 4. 
```

```python
class Employee:
    start_time = "9am"
    end_time = "5pm"

class admin(Employee):
    def __init__(self, role):
        self.role = role

class accountant(admin):
    def __init__(self, salary, role):
        super().__init__(role)
        self.salary = salary

emp1 = accountant(20_000 ,"CA")

print(emp1.start_time)
print(emp1.role)
print(emp1.end_time)
```

```python
# --------------------------------------Abstraction-------------------------------------------

# hiding internal details and showing only essential features

# to implement abstraction in real we use abstract class: also called blueprint of other classes

# to implement abstract class we need to import ABC and abstractmethod decorator from abc module
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

class lion(Animal):
    def make_sound(self, sound):
        self.sound = sound

class cow(Animal):
    def make_sound(self, sound):
        self.sound = sound

Lion = lion()
Cow = cow()

Lion.make_sound("roar!")
Cow.make_sound("moo!")
print(Lion.sound)
print(Cow.sound)
```

```python
#--------------------------------------Polymorphism---------------------------------------------

# multiple form, same name

# 1. function overriding

class Employee:
    def designation(self):
        print("Designation = Employee")

class Teacher(Employee):
    def designation(self):
        print("Designation = Teacher")

t1 = Teacher()

t1.designation()


# 2. duck typing

class CA:
    def designation(self):
        print("Designation = CA")

class Accountant:
    def designation(self):
        print("Designation = Accountant")

ca = CA()
acc = Accountant()

ca.designation()
acc.designation()
```

```python
class Student:
    __name = None
    __roll_no = None
    __marks = None

    def setter(self, name, rollNo, marks):
        if(name == ""):
            print("Name can't be empty")
            return 
        
        if(marks<0):
            print("Marks cant be negative")
            return 
        
        if(rollNo<1 or rollNo>100):
            print("Roll No cant be less than 1 or greater than 100")
            return
        
        self.__name = name
        self.__marks = marks
        self.__roll_no = rollNo

    def getter(self):
        print(f"Name of the student is {self.__name} with roll no {self.__roll_no} and marks {self.__marks}")


s1 = Student()
s1.setter("ayush", 35, 45)
s1.getter()
```

```python
class Person:
    def __init__(self, name, age = 21, address = "Jind"):
        self.name = name
        self.age = age
        self.address = address

p1 = Person("ayush")
p2 = Person("arun", 20)
p3 = Person("sunny", 23, "gohana")

print(f"Name of p1 is {p1.name} with age {p1.age} and address {p1.address}")
print(f"Name of p2 is {p2.name} with age {p2.age} and address {p2.address}")
print(f"Name of p3 is {p3.name} with age {p3.age} and address {p3.address}")
```
