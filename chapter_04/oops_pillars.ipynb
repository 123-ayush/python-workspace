```python
# -----------------------------------------Encapsulation-----------------------------------------

# -> binding data and methods together in a class/single unit is called encapsulation

# --------------------------------------data hiding---------------------------------------------

# 1. public attr. -> accessible everywhere in code
# 2. protected attr. -> to make any attr protected we need to put underscore before identifier name (inside class + subclass)
# 3. private attr. -> to make any attr protected we need to put double underscore before identifier name (inside class only)
# to access private attr. we need to use getter and setters
# to access private attr outside use this: 'obj._className__attrName'
```

```python
# ----------------------------------------Inheritance--------------------------------------------

# reusing methods/attr. of parent class in derived class

# Types:

    # 1. Single level
    # 2. Multilevel
    # 3. Multiple (class1, class2, ..........)
    # 4. 
```

```python
class Employee:
    start_time = "9am"
    end_time = "5pm"

class admin(Employee):
    def __init__(self, role):
        self.role = role

class accountant(admin):
    def __init__(self, salary, role):
        super().__init__(role)
        self.salary = salary

emp1 = accountant(20_000 ,"CA")

print(emp1.start_time)
print(emp1.role)
print(emp1.end_time)
```
