```python
# dictionary: key:pair values

dict = {
    "name": "ayush",
    "sub": ["science", "math"],
    "marks": [98, 99]
}

print(dict)

# dicts are mutable

dict["marks"][0] = 100

print(dict)
```

```python
# methods 

# 1. d.keys()
# 2. d.values()
# 3. d.items()
# 4. d.update(new_item)
# 5. d.get(val)

key = list(dict.keys())

print(key)

val = list(dict.values())

print(val)

print(dict.get('sub')) # if key not exist we dont get error we get none value

dict.update({
    "city" : "jind"
})

print(dict)
```
