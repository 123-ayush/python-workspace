```python
# tuple: immutable seq of values

tup = (1,2,3,4,"abc",10.56)

print(tup[5])

# to store single value in tup we need to put a comma after that element

tup1 = ("abc",)

print(type(tup1))

tup2 = ("abc")

print(type(tup2))

# slicing similar to list or string
```

```python
for val in tup:
    print(val)
```
