```python
# Lists: mutable seq of values

list = [1, 2, 23, 45, 67, "abc", 10.98]

print(list)

list[1] = 12

print(list)

# slicing is similar to string

new_list = list[-5:-1]

print(new_list)
```

```python
#   Methods in lists

# 1. l.append(val) -> add one element with val value at the end
# 2. l.insert(idx, val) -> insert element at idx
# 3. l.sort() -> sort the list
# 4. l.reverse() -> reverse the list order

# increasing order -> l.sort()
# decreasing order -> l.sort(reverse=True)
```

```python
# loops on lists

for val in list:
    print(val)

for i in range(len(list)):
    print(list[i])
```
