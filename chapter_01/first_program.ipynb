Character set:
    Letters: A-Z, a-z
    Numbers: 0-9
    Symbols: +, -, *, /, %, etc.
    Whitespaces: blank space, tab(\t), new line(\n);
    all ASCII(American Standard Code for Information Interchange) and Unicode chars as part of data and literals.

Variables:
    containers to store data.
    smallest container: 8 bits -> 1 byte

Identifiers:
    name of variables.
    Rules:
        1. we can use all english alphabets(case sensitive).
        2. all digit(0-9) and underscore( _ ).
        3. can't start name of variable with digit or any other symbol.
        4. Can't you keywords or reserved words as identifier like True, False, None, and, class, break, etc. 

Comments:
    1. when we want to write something for ourself not for python interpretor.
    2. starts with '#'.
    3. multiline comments: ''' '''

DataTypes:
    1. Integer(int) -> +ve, -ve, 0
    2. Float(float) -> all decimal number(+ve, -ve)
    3. Boolean(bool) -> True/False
    4. String(str) -> between "ayush", 'ayush', '''ayush'''
    5. None(NoneType) -> when we dont want to assign any value to any variable.

To get type of any variable, use print(type(var_name)).

We prefer snake_case in python.

To print any float value to a particular number of decimal use this:
