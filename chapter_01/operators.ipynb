Operators:
    1. Airthmetic
    2. Relational/Comparison
    3. Assignment
    4. Logical

Airthmetic Operators:
    1. Addition: '+'
    2. Subtraction: '-'
    3. Multiplication: '*'
    4. Division: '/'
    5. Modulus: '%'
    6. Power: '**'

```python
a = 10
b = 5

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
```

Relational/Comparison Operators:
    1. >, >=
    2. <, <=
    3. ==, !=

Return True/False value.

```python
a = 8
b = 8

print(a<b)
print(a<=b)
print(a>b)
print(a>=b)
print(a==b)
print(a!=b)
```

Assignment Operator:
    1. -=
    2. +=
    3. *=
    4. /=
    5. %=
    6. **=
    7. =

Logical Operators:
    1. and
    2. or
    3. not

```python
a = 5
b = 4

print(not(a>b))

print((a>b) and (a<b))

print((a>b) or (a==b))
```
