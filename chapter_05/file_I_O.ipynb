File Operations

```python
# Open operation

#   f = open("file_name/path", "modes")

f = open("example.txt", "r")
data = f.read()

print(data, type(data))

f.close()


f = open("example.txt", "w")
f.write("The ")
```

```python
# Modes:
    # 1. "r" -> read only mode [default]
        # a. f.read()
        # b. f.readline()
        # c. f.read(1)
    # 2. "w" -> write only mode but it overwrite the data in file
    # 3. "x" -> create new file & open for writing
    # 4. "a" -> append to write at the end of file
    # 5. "b" -> binary mode 
    # 6. "t" -> text mode [default]
    # 7. "+" -> open disk file for update(r & w)
```

```python
# ------------------------------------------With Keyword---------------------------------------------

# with the help of with keyword we dont need to explicitly close the file with close() method

with open("example.txt","r") as f:
    print(f.read())
```

```python
# ----------------------------------------Delete files--------------------------------------------

# os.remove("file_path")

import os
os.remove("example.txt")
```

```python
# -----------------------------------Word search acitivity----------------------------------------

word = "ayush"
flag = False
data = True

with open("example.txt", "r") as f:
    while data:
        data = f.readline()
        if(word in data):
            flag = True
            break
        

print(data,flag)
```
