```python
def sayHello():
    print("Hello World!")

sayHello()
```

```python
def add(a, b): # a,b is parameter
    return a+b

res = add(2,3) # 2,3 is argument

print(res)
```

```python
def avg(a,b,c=3):
    return (a+b+c/3)

a = avg(1,3)

print(f"{a:3.2f}")
```

Type of Function:
    1. Built-in function
    2. User-Defined function


```python
# LAMBDA Function
# we can have any number of parameter in it 
# but we can only have 1 expression

mul = lambda a,b: a*b

print(mul(2,3))
```
