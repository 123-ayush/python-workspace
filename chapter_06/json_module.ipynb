JSON -> JavaScript Object Notation

```python
import json

# -------------------------------json.loads()--------------------------------------------------

json_str = '{"name": "Ayush Ahlawat","age": 20}'

py_obj = json.loads(json_str)

print(type(py_obj), py_obj)
```

```python
# -------------------------------json.dumps()--------------------------------------------------

py_str = {
    "name": "Ayush Ahlawat",
    "age": True
}

json_obj = json.dumps(py_str)

print(type(json_obj), json_obj)
```

```python
# -------------------------------json.load()--------------------------------------------------

with open("data.json", "r") as f:
    json_obj = json.load(f)
    print(json_obj)
```

```python
# -------------------------------json.dump()--------------------------------------------------

new_data = {
    "name": "arun",
    "age": 20,
    "isStudent": True
}

with open("data.json", "w") as f:
    json.dump(new_data, f, indent=4, sort_keys=True)
```
