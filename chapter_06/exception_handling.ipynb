```python
# -------------------------------------------Exceptional Handling---------------------------------------

# try -> part of code which is possibly can through an error or give an error
# except -> handle exception
# else -> if no exception occur
# finally -> run everytime exception come or not


try:
    deno = int(input("Enter number: "))
    res = 25/deno
    print(res)

except ZeroDivisionError:
    print("Denominator cant be zero.")

else:
    print("No exception occur")

finally:
    print("Final block executed.")
```
